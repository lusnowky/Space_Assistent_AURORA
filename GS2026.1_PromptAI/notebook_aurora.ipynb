```python
from kaggle_secrets import UserSecretsClient
from huggingface_hub import InferenceClient

user_secrets = UserSecretsClient()

HF_TOKEN = user_secrets.get_secret("HUGGING_FACE_API_KEY")

modelo = "meta-llama/Meta-Llama-3-8B-Instruct"

client = InferenceClient(api_key=HF_TOKEN)

print("Modelo configurado com sucesso!")
```

```python
print("===================================== AURORA =====================================")
print("Olá, eu sou AURORA, sua assistente de monitoramento espacial")
print("Vamos analisar os sistemas da nave para garantir uma viagem segura e eficiente!")
print("\n")

temperatura = float(input("Temperatura da nave (°C): "))
energia = float(input("Nível de energia (%): "))
comunicacao = input("Comunicação funcionando? (sim/não): ")
```

```python
# Temperatura

if temperatura > 80:
    status_temperatura = "Alerta de Superaquecimento"
else:
    status_temperatura = "Temperatura Normal"

# Comunicação

if comunicacao.lower() == "sim":
    status_comunicacao = "Comunicação Estável"
else:
    status_comunicacao = "Falha na Comunicação"

# Energia

if energia < 20:
    status_energia = "Alerta de Baixa Energia"
else:
    status_energia = "Nível de Energia Adequado"

print("\n=== ANÁLISE PRELIMINAR ===")
print(status_temperatura)
print(status_comunicacao)
print(status_energia)
```

```python
prompt = f"""
Você é uma inteligência artificial chamada AURORA especializada em monitoramento de missões espaciais.

Analise os dados abaixo.

DADOS DA MISSÃO

Temperatura: {temperatura} °C

Energia: {energia} %

Comunicação: {comunicacao}

RESULTADO DA ANÁLISE DE SEGURANÇA

Temperatura:
{status_temperatura}
Comunicação:
{status_comunicacao}
Energia:
{status_energia}

REGRAS OPERACIONAIS DE SEGURANÇA

- Temperaturas acima de 80°C representam risco de superaquecimento.
- Energia abaixo de 20% representa risco operacional.
- Falhas de comunicação representam perda de contato com a base.

Sua tarefa é agir como uma especialista em controle de missão espacial.

Identificar se, mesmo que sem falhas, os valores de Temperatura e Energia estão próximos ao limite seguro (presente nas REGRAS OPERACIONAIS DE SEGURANÇA), e se estiverem alertar na PREVISÃO

Responda obrigatoriamente utilizando a seguinte estrutura:

STATUS GERAL:

FALHAS DETECTADAS: (Se basear caso alguma das variáveis a seguir indicar falha e/ou alerta "status_temperatura", "status_comunicacao" e "status_energia")

NÍVEL DE RISCO:

PREVISÃO:

RECOMENDAÇÕES:

Explique de forma técnica e profissional.
"""
```

```python
response = client.chat.completions.create(
    model="meta-llama/Llama-3.1-8B-Instruct",
    messages=[
        {
            "role": "system",
            "content": """
Você é uma inteligência artificial especialista em monitoramento de missões espaciais.

Suas responsabilidades são:
- Analisar os dados da missão.
- Detectar falhas operacionais.
- Avaliar riscos.
- Prever possíveis problemas futuros.
- Recomendar ações corretivas.

Utilize linguagem técnica, objetiva e profissional.
"""
        },
        {
            "role": "user",
            "content": prompt
        }
    ],
    max_tokens=500,
    temperature=0.4
)

print("\n")
print("=" * 28)
print("RELATÓRIO GERADO PELA AURORA")
print("=" * 28)
print("\n")

print(response.choices[0].message.content)
```
